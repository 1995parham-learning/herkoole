# Traveling Salesman Problem

```python
from main import tsp
from ea import EvolutionaryAlgorithm

import matplotlib.pyplot as plt
```

```python
with open('../tsp_data.txt') as f:
    m = tsp(f)
alg = EvolutionaryAlgorithm(10, 20, 8000, m, window_size=1000, threshold=-1)
ch = alg.run()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

ax.plot(alg.average_fitness, '-o')
ax.set_xlabel('Generation')
ax.set_ylabel('Average Fitness')
```

```python
xs = []
ys = []
for gene in ch.genes:
    city = m.cities[gene]
    xs.append(city.x)
    ys.append(city.y)
    
fig, ax = plt.subplots(figsize=(25, 25))

ax.plot(ys, xs, markeredgecolor='red', marker='o', markersize=4)
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
