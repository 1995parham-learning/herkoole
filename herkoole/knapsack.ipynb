# Knapsack Problem

```python
from main import knapsack
from ea import EvolutionaryAlgorithm

import matplotlib.pyplot as plt
```

## Knapsack 1

```python
with open('../knapsack_1.txt') as f:
    m = knapsack(f)
alg = EvolutionaryAlgorithm(10, 20, 1000, m, window_size=1000, threshold=-1)
print(alg.run())
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

ax.plot(alg.average_fitness, '-o')
ax.set_xlabel('Generation')
ax.set_ylabel('Average Fitness')
```

Here we cannot find a feasible solution. The final fitness and average fitness is not above 1 so there is no feasible soltion to given knapsack problem.
